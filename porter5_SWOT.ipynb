```python
import matplotlib.pyplot as plt
import numpy as np

def porters_five_forces(company, industry, forces=None, save=False):
    """
    Generate a radar chart for Porter's Five Forces Analysis.

    Parameters:
        company (str): Company name
        industry (str): Industry name
        forces (dict): Optional custom scores (1 to 5) for each force
        save (bool): Whether to save the chart as PNG
    """
    # Default scores (can be overridden)
    default_forces = {
        'Threat of New Entrants': 3,
        'Bargaining Power of Suppliers': 3,
        'Bargaining Power of Buyers': 3,
        'Threat of Substitutes': 3,
        'Industry Rivalry': 3
    }

    if forces:
        default_forces.update(forces)

    categories = list(default_forces.keys())
    values = list(default_forces.values())
    values += values[:1]  # repeat first value to close the radar chart
    categories += categories[:1]

    angles = [n / float(len(categories)) * 2 * np.pi for n in range(len(categories))]

    # Create radar plot
    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
    ax.plot(angles, values, linewidth=2, linestyle='solid', label=company)
    ax.fill(angles, values, alpha=0.25)

    # Customization
    ax.set_xticks(angles)
    ax.set_xticklabels(categories, fontsize=10)
    ax.set_yticklabels([])
    ax.set_ylim(0, 5)
    ax.set_title(f"Porterâ€™s Five Forces:\n{company} ({industry})", size=14, weight='bold', pad=20)
    plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))

    if save:
        filename = f"{company.replace(' ', '_').lower()}_porters_5_forces.png"
        plt.savefig(filename, bbox_inches='tight')
        print(f"âœ… Saved: {filename}")

    plt.show()

# Example Usage:
nvidia_forces = {
    'Threat of New Entrants': 2,
    'Bargaining Power of Suppliers': 4,
    'Bargaining Power of Buyers': 3,
    'Threat of Substitutes': 3,
    'Industry Rivalry': 5
}

porters_five_forces("NVIDIA Corporation", "Semiconductors", forces=nvidia_forces, save=True)
```

```python
import matplotlib.pyplot as plt

def swot_analysis(company, strengths, weaknesses, opportunities, threats, save=False):
    """
    Visualizes a SWOT analysis in quadrant format.

    Parameters:
        company (str): Name of the company.
        strengths, weaknesses, opportunities, threats (list): SWOT elements.
        save (bool): If True, saves the plot as PNG.
    """
    fig, ax = plt.subplots(figsize=(10, 6))
    plt.axis('off')

    ax.set_title(f"SWOT Analysis: {company}", fontsize=16, fontweight='bold', pad=20)

    # Box styling
    box_props = dict(boxstyle='round,pad=0.4', facecolor='#f7f7f7', edgecolor='gray')

    # Top Left: Strengths
    ax.text(0.01, 0.85, "Strengths", fontsize=14, color='green', fontweight='bold')
    ax.text(0.01, 0.62, '\n'.join(f"- {s}" for s in strengths), fontsize=11, bbox=box_props)

    # Top Right: Weaknesses
    ax.text(0.55, 0.85, "Weaknesses", fontsize=14, color='red', fontweight='bold')
    ax.text(0.55, 0.62, '\n'.join(f"- {w}" for w in weaknesses), fontsize=11, bbox=box_props)

    # Bottom Left: Opportunities
    ax.text(0.01, 0.45, "Opportunities", fontsize=14, color='blue', fontweight='bold')
    ax.text(0.01, 0.22, '\n'.join(f"- {o}" for o in opportunities), fontsize=11, bbox=box_props)

    # Bottom Right: Threats
    ax.text(0.55, 0.45, "Threats", fontsize=14, color='darkorange', fontweight='bold')
    ax.text(0.55, 0.22, '\n'.join(f"- {t}" for t in threats), fontsize=11, bbox=box_props)

    plt.tight_layout()

    if save:
        filename = f"{company.replace(' ', '_').lower()}_swot_analysis.png"
        plt.savefig(filename, bbox_inches='tight')
        print(f"âœ… Saved: {filename}")

    plt.show()

# Example: NVIDIA SWOT
strengths = [
    "Leader in AI and graphics chips",
    "Strong brand and innovation",
    "Robust R&D capabilities",
    "High profitability margins"
]

weaknesses = [
    "Heavy reliance on TSMC for manufacturing",
    "Limited diversification beyond GPU & AI",
    "High product cost"
]

opportunities = [
    "Expansion into automotive AI and self-driving",
    "Increasing global demand for AI and cloud",
    "Growth in data center and HPC segments"
]

threats = [
    "Strong competition from AMD, Intel, custom chips",
    "Geopolitical tension (US-China, Taiwan)",
    "Chip supply constraints",
    "Regulatory scrutiny"
]

swot_analysis("NVIDIA Corporation", strengths, weaknesses, opportunities, threats, save=True)
```

```python
# from FinGPT # FinGPT Agent
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
import re

# Initialize FinGPT
agent = FinGPT(verbose=False)

# 1ï¸âƒ£ Fetch basic info & summary
def fetch_company_info(ticker):
    info = yf.Ticker(ticker).info
    return {
        "name": info.get("longName", ticker),
        "ticker": ticker,
        "sector": info.get("sector", "N/A"),
        "industry": info.get("industry", "N/A"),
        "summary": info.get("longBusinessSummary", "")
    }

# 2ï¸âƒ£ Prompt FinGPT to analyze Porterâ€™s Five Forces
def porter_with_fingpt(company):
    prompt = f"""
You are a seasoned financial analyst. Provide a detailed Porter's Five Forces analysis for {company['name']} ({company['ticker']}) in its industry.

Include:
1. Competitive Rivalry
2. Threat of New Entrants
3. Bargaining Power of Suppliers
4. Bargaining Power of Buyers
5. Threat of Substitutes

Add a rating (1â€‘5) after each section like: Score: 4
"""
    return agent.chain.run(prompt)

# 3ï¸âƒ£ Parse the output into labels + numeric scores
def parse_porter(text):
    labels = []
    scores = []
    lines = text.splitlines()
    for line in lines:
        m = re.match(r"^\d\.\s*(.+):\s*(.+?)Score:\s*([1-5])", line)
        if m:
            labels.append(m.group(1).strip())
            scores.append(int(m.group(3)))
    return labels, scores, text

# 4ï¸âƒ£ Plot radar chart
def plot_radar(labels, scores, title):
    labels_ext = labels + [labels[0]]
    scores_ext = scores + [scores[0]]
    angles = np.linspace(0, 2*np.pi, len(labels_ext), endpoint=False).tolist()
    fig, ax = plt.subplots(figsize=(6,6), subplot_kw={'polar': True})
    ax.plot(angles, scores_ext, color='teal', linewidth=2)
    ax.fill(angles, scores_ext, alpha=0.3, color='teal')
    ax.set_xticks(angles)
    ax.set_xticklabels(labels_ext, fontsize=10)
    ax.set_yticks([1,2,3,4,5])
    ax.set_title(title, fontsize=14)
    plt.show()

# 5ï¸âƒ£ Full workflow
def analyze_porter_fingpt(ticker):
    company = fetch_company_info(ticker)
    print(f"ðŸ” Fetching & analyzing: {company['name']} ({ticker})\n")
    output = porter_with_fingpt(company)
    labels, scores, raw = parse_porter(output)
    print("ðŸ“Š Porterâ€™s Five Forces Analysis:\n")
    print(raw.strip(), "\n")
    if labels and scores:
        plot_radar(labels, scores, company['name'])
    else:
        print("âš ï¸ Unable to parse structured scores. Here's raw output.")

# ðŸš€ Example
analyze_porter_fingpt("NVDA")
```
